Store procedure:-
in this data set we have created 3 store procedure  name



1.	getbasic:- with the help of this procedure we can find all the basic details of particular drug by its drug id .
and the mysql query for this procedure is:-


CREATE DEFINER=`root`@`localhost` PROCEDURE `getbasic`(id varchar(5))
BEGIN
select * from drugdotcom.drugbasic where D_id=id;
END


2.	getname_imprint:- with the help of this procedure we can find drug name and its imprint.
and the mysql query for this procedure is:-
CREATE DEFINER=`root`@`localhost` PROCEDURE `getname_imprint`()
BEGIN
select drugbasic.Drug_name, pill_identifier.pill_imprint from drugdotcom.drugbasic
inner join drugdotcom.pill_identifier
on drugbasic.D_id=pill_identifier.D_id;
END

# Trigger:-

We have created trigger (druginteraction)  in drug interaction table in which trigger activates whenever a new row is inserted into the table.
And the query for this trigger is :-


CREATE DEFINER=`root`@`localhost` TRIGGER `druginteraction_BEFORE_INSERT` BEFORE INSERT ON `druginteraction` FOR EACH ROW BEGIN
if new.total !=new.moderate+new.major+new.minor then
set new.total=new.moderate+new.major+new.minor;
elseifnew.total = new.moderate+new.major+new.minor
then
set new.total=new.total;
end if;
END



Call trigger:- by this trigger when we insert value of major ,moderate and major value if we insert wrong value of its total by this trigger this will sum all the value and give exect total of table.

# materialized view:-

we have created materialized view for the table ddt by this materialized view we can view original table anytime weather it is  modified ,update or delete.

Query for materialized view:-
CREATE
ALGORITHM = UNDEFINED
DEFINER = `root`@`localhost`
SQL SECURITY DEFINER
create materialized view `drugdotcom`.`details` AS
SELECT
`drugdotcom`.`ddt`.`Dt_id` AS `Dt_id`,
`drugdotcom`.`ddt`.`drugName` AS `drugName`,
`drugdotcom`.`ddt`.`url` AS `url`,
`drugdotcom`.`ddt`.`description` AS `description`
FROM
`drugdotcom`.`ddt`
