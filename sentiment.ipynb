```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt

```

```python

df=pd.read_csv("drugsComTest_raw.csv")
df=pd.DataFrame(df)
df
```

```python

review=df['review']
```

```python

positive=0
negative=0
neutral=0
rev=[]
for r in review:
  analysis=TextBlob(r)
  print(analysis.sentiment)
  if analysis.sentiment.polarity>0:
    print("positive")
    rev.append('positive')
    positive=positive+1
  elif analysis.sentiment.polarity==0:
    print("neutral") 
    rev.append('neutral')
    neutral=neutral+1
  else:
     print("negative")
     rev.append('negative')
     negative=negative+1  
```
